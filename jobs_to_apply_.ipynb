```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://realpython.github.io/fake-jobs/"
```

```python
r=requests.get(url)
r
```

```python
r.content
```

```python
soup=BeautifulSoup(r.content,"html")
soup
```

```python
soup.find("h2",attrs={"class":"title is-5"}).text
```

```python
soup.find("h3",attrs={"class":"subtitle is-6 company"}).text
```

```python
soup.find("p",attrs={"class":"location"}).text.strip()
```

```python
soup.find("p",attrs={"class":"is-small has-text-grey"}).text.strip()
```

```python
title=soup.find_all("h2",attrs={"class":"title is-5"})
tit=[]
for i in title:
  tit.append(i.text)
tit
```

```python
sub_title=soup.find_all("h3",attrs={"class":"subtitle is-6 company"})
sub=[]
for i in sub_title:
  sub.append(i.text)
sub
```

```python
location=soup.find_all("p",attrs={"class":"location"})
loc=[]
for i in location:
  loc.append(i.text.strip())
loc
```

```python
date_posted=soup.find_all("p",attrs={"class":"is-small has-text-grey"})
date=[]
for i in date_posted:
  date.append(i.text.strip())
date
```

```python
data=pd.DataFrame({"job title":tit,
                   "sub title":sub,
                   "location":loc,
                   "date posted":date})
data
```

```python
data.to_csv("jobs-to-apply.csv")
```
